```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df['target']=cancer['target']
```

```python
df.head(50)
```

```python
df.info()
```

```python
sns.pairplot(df,hue='target',vars=['mean radius','mean texture','mean perimeter','mean smoothness'])
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['target'],axis=1)
y=df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
pred=svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
cm
#sns.heatmap(cm,annot=True)
```

```python
x_min=X_train.min()
```

```python
range_train=(X_train-x_min).max()
```

```python
X_scaled=(X_train-x_min)/range_train
X_test_scaled=(X_test-x_min)/range_train
```

```python
svc.fit(X_scaled,y_train)
```

```python
predict=svc.predict(X_test_scaled)
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(classification_report(y_test,predict))
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_scaled,y_train)
```

```python
grid.best_params_
```

```python
best_pre=grid.predict(X_test_scaled)
```

```python
print(confusion_matrix(y_test,best_pre))
```

```python
print(classification_report(y_test,best_pre))
```
